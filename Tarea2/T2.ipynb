```python
df_chi = pd.read_excel('MigrantesChile (2005-2016).xlsx',
                      sheet_name='Migrantes',
                      skipfooter=3)

print ('Data read into a pandas dataframe!')
```

```python
df_chi.loc[df_chi['Continent'] == 'America']
df_chi.drop(['ID Continent','ID Country'], axis=1, inplace=True)
```

```python
df_chi
```

```python
df_chi.set_index('Country', inplace=True)


```

```python
df_pcb = df_chi.loc[['Peru', 'Colombia','Argentina', 'Bolivia' ,'Brazil', 'Ecuador', 'Paraguay','Trinidad and Tobago', 'Uruguay', 'Venezuela', 'Guyana', 'Suriname'], :]
```

```python
df_pcb
```

```python
total_values = __builtins__.sum(df_pcb[2012])
valores = [valor for valor in df_pcb[2012]]
print(valores)
category_proportions = [(float(value) / total_values) for value in df_pcb[2012]]

# imprimir proporciones
for i, proportion in enumerate(category_proportions):
    print (df_pcb.index.values[i] + ': ' + str(proportion))
```

```python
width = 40 # ancho del of chart
height = 10 # altura del chart

total_num_tiles = width * height # número total de celdas

print ('Número total de celdas: ', total_num_tiles)
```

```python
# calcula la cantidad de celdas por categoría
tiles_per_category = [int(round(proportion * total_num_tiles)) for proportion in category_proportions]
tiles_per_category1 = [(round(proportion * 100)) for proportion in category_proportions]
print(tiles_per_category1)
# imprime número de celdas por categoría
for i, j, k  in zip(df_pcb.index,tiles_per_category, tiles_per_category1):
    print (f'{i} : {str(j)} --> {str(k)} %')
#for i, tiles1 in enumerate(tiles_per_category1):
#    print (df_pcb.index.values[i] + ': ' + str(tiles1))
```
